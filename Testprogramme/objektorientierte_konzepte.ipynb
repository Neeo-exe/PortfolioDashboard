```python
class Pruefungsleistung:
    def __init__(self, modul, note, bestanden):
        self.modul = modul
        self.note = note
        self.bestanden = bestanden

pruefung = Pruefungsleistung("Betriebssysteme, Rechnernetze und verteilte Systeme", 3.3, True)

print("Modul:", pruefung.modul)
print("Note:", pruefung.note)
print("Bestanden:", pruefung.bestanden)
```

```python
from enum import Enum

class ModulStatus(Enum):
    OFFEN = 1
    IN_BEARBEITUNG = 2
    BESTANDEN = 3

status = ModulStatus.BESTANDEN

print("Status:", status)
```

```python
class Studiengang:
    def __init__(self, name):
        self.name = name

class StudierendePerson:
    def __init__(self, name, studiengang):
        self.name = name
        self.studiengang = studiengang

studiengang = Studiengang("Cyber Security")
person = StudierendePerson("Kevin", studiengang)

print("Name:", person.name)
print("Studiengang:", person.studiengang.name)
```

```python
class Modul:
    def __init__(self, titel):
        self.titel = titel

class Semester:
    def __init__(self):
        self.module = []

semester = Semester()

semester.module.append(Modul("Betriebssysteme, Rechnernetze und verteilte Systeme"))

print(semester.module[0].titel)
```
